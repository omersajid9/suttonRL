```python
import numpy as np
from collections import defaultdict
```

```python

class Board:
    def __init__(self, rows, cols):
        self.board = [[0] * cols for _ in range(rows)]

    def to_string(self):
        rows_array = []

        for rows in self.board:
            string_row = [str(e) for e in rows]
            col_joined = ",".join(string_row)
            rows_array.append(col_joined)
        return "|".join(rows_array)
    
    def from_string(self, string_board):
        board = []

        rows = string_board.split("|")
        for row in rows:
            col_string = row.split(",")
            col = [int(e) for e in col_string]
            board.append(col)
        self.board = board

    def move(self, player, pos):
        x, y = pos
        self.board[x][y] = 1 if player == 1 else -1
```

```python
r, c  = 3, 3
b = Board(r, c)
b.move(1, (1, 1))
b.from_string(b.to_string())
b.board
```
